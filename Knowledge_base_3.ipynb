```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_chroma import Chroma
from langchain_core.prompts import ChatPromptTemplate
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
from langchain_classic.memory import ConversationBufferMemory
from langchain_classic.chains import ConversationalRetrievalChain
```

```python
import getpass
import os
os.environ["GOOGLE_API_KEY"]=getpass.getpass("Enter API key:")
```

```python
#llm
llm=ChatGoogleGenerativeAI(model="gemini-2.0-flash",temperature=0.3)# temp controls randomness like it is good for creative ans.
```

```python
#pdf_loader
pdf_file=["/content/meditations.pdf","/content/Life-Changing Foods PDF.pdf"]
```

```python
all_docs=[]
for file in pdf_file:
  loader=PyPDFLoader(file)
  all_docs.extend(loader.load())#combine them into one list
print(f"✅ Loaded {len(all_docs)} total pages from {len(pdf_file)} PDFs.")
```

```python
#spliting
spliter=RecursiveCharacterTextSplitter(chunk_size=1000,chunk_overlap=200)
chunk=spliter.split_documents(all_docs)
```

```python
#embedings
embeding=HuggingFaceEmbeddings(
    model_name="BAAI/bge-base-en-v1.5",
    model_kwargs={"device":"cpu"},
    encode_kwargs={"normalize_embeddings":True}
    )
```

```python
#vectorstore
vectorstore=Chroma.from_documents(chunk,embeding)
```

```python
#retriver
retriever=vectorstore.as_retriever(search_kwargs={"k":3})
```

```python
memory = ConversationBufferMemory(memory_key="chat_history", return_messages=True)
```

```python
chain = ConversationalRetrievalChain.from_llm(
    llm=llm,
    retriever=retriever,
    memory=memory,
    verbose=True
)
```

```python
# 9️⃣ Chat loop
print("exit to quit.\n")
while True:
    query = input("You: ")
    if query.lower() in ["exit", "quit"]:
        print("Goodbye")
        break
    response = chain.run(query)
    print("Bot:", response)
```
