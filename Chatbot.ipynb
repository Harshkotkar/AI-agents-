```python
import langchain_core
```

```python
from langchain_classic.memory import ConversationBufferMemory
from langchain_classic.chains import ConversationChain
```

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_google_genai import ChatGoogleGenerativeAI
```

```python
import getpass
import os
os.environ["GOOGLE_API_KEY"]=getpass.getpass("Enter google api key:")
```

```python
llm=ChatGoogleGenerativeAI(model="gemini-2.0-flash-exp",temperature=0.1)
```

```python
memory = ConversationBufferMemory(return_messages=True)
```

```python
prompt = ChatPromptTemplate.from_template(
    """You are a helpful assistant that provides healthy lifestyle and meditation advice. provide it into easy to understandable way. use points

Chat history:
{history}

User: {input}
Assistant:"""
)
```

```python
# chain=llm|prompt|memory
# query="hey i have migrane what do i do?"
# response=chain.invoke(query)
```

```python
chain=ConversationChain(llm=llm,
                        prompt=prompt,
                        memory=memory,
                        verbose=True)
while True:
  query=input("user: ")
  if query.lower()in["exit","quit"]:
    print("Bye take care")
    break
  response=chain.invoke(query)
  print("Chatbot:",response)
```
