```python
from pydantic import BaseModel
```

```python
class human(BaseModel):
  f_name:str
  l_name:str
  age:int
spiderman=human(f_name="peter",l_name="parker",age=18)
print(spiderman)
```

```python
from pydantic import BaseModel

class address(BaseModel):
  street:str
  city:str
  zip:int

class customer(BaseModel):
  name:str
  id:int
  address:address # nested looping

#create a customer with nested address
cust1=customer(
    name="tony",
    id=98711,
    address={"street":"queens street","city":"newyork","zip":1000}
)
print(cust1)
```
