```python
# Just store messages in a list
history = []

def chat(user_input):
    history.append(f"User: {user_input}")
    context = "\n".join(history)
    prompt = f"""You are a friendly chatbot. Continue the conversation naturally.
{context}
AI:"""

    response = llm.invoke(prompt)
    reply = response.content.strip()
    history.append(f"AI: {reply}")
    return reply
```

```python
%%writefile app.py
import streamlit as st
from langchain_google_genai import ChatGoogleGenerativeAI

st.set_page_config(page_title="Friendly  Chatbot", page_icon="üí¨")

llm=llm = ChatGoogleGenerativeAI(model="gemini-2.0-flash-exp", temperature=0.2)

if "history" not in st.session_state:
    st.session_state.history = []

def chat(user_input):
    st.session_state.history.append(f"User: {user_input}")
    context = "\n".join(st.session_state.history)
    prompt = f"""You are a friendly chatbot. Continue the conversation naturally.
{context}
AI:"""
    response = llm.invoke(prompt)
    reply = response.content.strip()
    st.session_state.history.append(f"AI: {reply}")
    return reply

st.title("üí¨ Friendly Chatbot")
st.write("Say hi! The bot responds in a friendly, conversational tone.")

user_input = st.chat_input("Type your message here...")
if user_input:
    reply = chat(user_input)

for message in st.session_state.history:
    if message.startswith("User:"):
        st.chat_message("user").write(message.replace("User:", "").strip())
    else:
        st.chat_message("assistant").write(message.replace("AI:", "").strip())

if st.button("üóëÔ∏è Clear Chat"):
    st.session_state.history = []
    st.experimental_rerun()
```
