```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import PromptTemplate
```

```python
from langchain_classic.chains import LLMChain
from langchain_classic.agents import initialize_agent, Tool
```

```python
import json
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)
```

```python
import os
import getpass
os.environ["GOOGLE_API_KEY"]=getpass.getpass("enter api key")
```

```python
llm=ChatGoogleGenerativeAI(model="gemini-2.0-flash-exp",temperature=0.1)
```

```python
from ddgs import DDGS

def web_search(query: str) -> str:
  # web search using DuckDuckGo (ddgs) and return top results.
    try:
        results = []
        with DDGS() as ddgs:
            for r in ddgs.text(query, region='wt-wt', safesearch='moderate', timelimit='m', max_results=5):
                title = r.get("title", "")
                body = r.get("body", "")
                link = r.get("href", "")
                results.append(f"Title: {title}\nSnippet: {body}\nURL: {link}")
        return "\n\n".join(results) if results else "No results found."
    except Exception as e:
        return f"Search failed: {e}"
```

```python
print(web_search("Cancer treatment in 2025"))
```

```python
def summarizer(text):
  prompt=PromptTemplate(
      input_vaiables=["text"],
      template="Consider you a proffectional summarizer. Summarize the following content in the proffection language without lossing the context of the content and the main idea of the content. Summarize the content in 5-6 bullet points  summary:\n\n{text} "
  )
  chain=LLMChain(llm=llm,prompt=prompt)
  result=chain.run({"text":text})
  return result
```

```python
#save notes
NOTES_FILE = "notes.json"
def save_notes(note_text: str) -> str:
    note = {"text": note_text}
    try:
        if os.path.exists(NOTES_FILE):
            with open(NOTES_FILE, "r", encoding="utf-8") as f:
                notes = json.load(f)
        else:
            notes = []
        notes.append(note)
        with open(NOTES_FILE, "w", encoding="utf-8") as f:
            json.dump(notes, f, indent=2)
        return f"Note saved ({len(notes)} total)."
    except Exception as e:
        return f"Failed to save note: {e}"
```

```python
def email_drafter(context):
# Draft an email based on context. Doesn't send — just drafts.


    prompt = PromptTemplate(
        input_variables=["context"],
        template="Draft a professional email to send the following summary to my team:\n\n{context}\n\nInclude subject line and 3 short bullet points."
    )
    chain = LLMChain(llm=llm, prompt=prompt)
    return chain.run({"context": context})
```

```python
tools = [
    Tool(name="web_search", func=web_search, description="Search the web for information and return ."),
    Tool(name="summarizer", func=summarizer, description="Summarize long texts or articles."),
    Tool(name="save_notes", func=save_notes, description="Save a note to local storage."),
    Tool(name="email_drafter", func=email_drafter, description="Draft an email from provided context (does not send).")
]
```

```python
agent = initialize_agent(
    tools=tools,
    llm=llm,
    agent_type="zero-shot-react-description",
    verbose=True
)
```

```python
def main():
    print("Chat-Based Task Assistant (type 'exit' to quit)\n")
    while True:
        user_input = input("You: ").strip()
        if not user_input:
            continue
        if user_input.lower() in ("exit", "quit"):
            break

        # Run the agent
        try:
            result = agent.run(user_input)
            print("\nAgent:", result)
        except Exception as e:
            print("Error:", e)

if __name__ == "__main__":
    main()
```
